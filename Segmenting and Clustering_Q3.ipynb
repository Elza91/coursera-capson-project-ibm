```python
# importar la libreria requests para descargar los datos de formato HTML
import requests
# Requests cuenta con una forma muy simple de implementación, un ejemplo de una consulta GET sería:
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
print('all good')
```

__En este tutorial vamos ha "scrapear" (o a obtener los datos) de los articulos y tutoriales de esta página web, utilizando la librería "BeautifulSoup" (para Python) que es una libreria que nos facilita muchisimo la labor de scrapeo.

```python
from bs4 import BeautifulSoup as bs
soup = bs(url,'lxml')
print(soup.prettify())
```

```python
# Si miramos todos los resultados anteriores vemos que solo nos interesa la class='wikitable sortable' y lo guardo en My_table
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
print(My_table.tr.text)
```

```python
# ahora hacemos la cabecera del dataframe
headers="Postcode,Borough,Neighbourhood"
headers
```

```python
# si vemos anteriormente hay dos estructuras <tr> ... <td> <td> ... <tr>:
table0=""
for tr in My_table.find_all('tr'):
    row0=""
    for tds in tr.find_all('td'):
        row0=row0+","+tds.text
    table0=table0+row0[1:]
print(table0)
```

```python
# una vez lo tengamos así, ahora lo pasamos a formato .csv para mejorar la comprensión y ponemos que escribirá en él = 'WB'
file=open("toronto.csv","wb")
```

```python
file.write(bytes(table0,encoding="ascii",errors="ignore"))
```

__a.1.The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood__

```python
# lo pasamos a dataframe con pandas 
import pandas as pd
df = pd.read_csv('toronto.csv',header=None)
df.columns=["Postalcode","Borough","Neighbourhood"]
# para ver el tamaño del dataframe
df.shape
```

```python
df.head(5)
```

__a.2.Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.__

```python
# Ver los índices que tienen 'Borough' no asignado a ningun valor y borrarlo con drop
indexNames = df[ df['Borough'] =='Not assigned'].index

df.drop(indexNames , inplace=True)
```

```python
df.head(10).transpose()
# aqui vemos que no todas las filas saltan
```

__a.3.If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.__

```python
df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
df.head(10).transpose()
```

__a.4.More than one neighborhood can exist in one postal code area.__

```python
result = df.groupby(['Postalcode','Borough'], sort=False).agg( ', '.join)
df_new=result.reset_index()
df_new.head(8).transpose()
```

```python
# conviente una address en valores de latitud y longitud. 
from geopy.geocoders import Nominatim

import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude_toronto = location.latitude
longitude_toronto = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))

```

```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# add markers to map
for lat, lng, borough, Neighbourhood in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Borough'], Toronto_df['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
